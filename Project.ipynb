```python
import pandas as pd
```

```python
data_saver=pd.read_csv('/content/sales_data.csv')
data_saver
```

```python
df=pd.DataFrame(data_saver)
df=df.drop_duplicates()
df
```

```python

print("Missing values per column:")
print(df.isnull().sum())
missing_data = df[df.isnull().any(axis=1)]
print(missing_data)


print("\nRows with invalid price (negative):")
print(df[df['price'] < 0])

print("\nRows with invalid quantity (negative):")
print(df[df['quantity'] < 0])
```

```python
df['revenue'] = df['revenue'].fillna(df['price'] * df['quantity'])
df['price'] = df['price'].fillna(df['revenue'] / df['quantity'])
df['quantity'] = df['quantity'].fillna(df['revenue'] / df['price'])
print(df)
missing_data = df[df.isnull().any(axis=1)]
print(missing_data)
```

```python
df['quantity'] = df['quantity'].astype(int)
print(df[['price', 'quantity', 'revenue']].dtypes)
```

```python
df['date'] = pd.to_datetime(df['date'])
df['quarter'] = df['date'].dt.quarter.apply(lambda x: f'Q{x}')
print(df)
```

```python
df['price_check'] = (df['price'] * df['quantity'] == df['revenue']).astype(int)
print(df)
```

```python

print("Missing values per column:")
print(df.isnull().sum())
missing_data = df[df.isnull().any(axis=1)]
print(missing_data)


print("\nRows with invalid price (negative):")
print(df[df['price'] < 0])

print("\nRows with invalid quantity (negative):")
print(df[df['quantity'] < 0])
```

```python
print(df[df['price_check'] == 0])
```

```python
import sqlite3

conn = sqlite3.connect('sales_data')
df.to_sql('sales_table', conn, if_exists='replace')
print("DataFrame saved to SQLite database successfully.")


read_df = pd.read_sql('SELECT * FROM sales_table', conn)
print("\nData read back from SQLite:")
print(read_df)

```

```python
print("Data read back from SQLite:")
read_df = pd.read_sql('SELECT SUM(revenue) FROM sales_table', conn)

print(read_df)
```

```python
print("Data read back:")
read_df = pd.read_sql('SELECT SUM(quantity) FROM sales_table', conn)
print(read_df)
```

```python
print("Data read back:")
read_df = pd.read_sql('SELECT AVG(Price) FROM sales_table', conn)
print(read_df)
```

```python
print("Data read back:")
read_df = pd.read_sql('SELECT product, MAX(Revenue) FROM sales_table Group by product Order by Revenue DESC limit 1 ', conn)
print(read_df)
```

```python
print("Data read back:")
read_df = pd.read_sql('SELECT Category, SUM(Revenue) FROM sales_table Group by Category Order by Revenue DESC limit 1', conn)
print(read_df)
```

```python
print("Data read back:")
read_df = pd.read_sql('SELECT quarter, SUM(Revenue) FROM sales_table Group by quarter ', conn)
print(read_df)
conn.close()
```
